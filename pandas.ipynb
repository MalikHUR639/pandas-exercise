# Concate dataframes

```python
import pandas as pd
df=pd.DataFrame({
    'city':["lahore","karachi","lyaa"],
    'temp':[24,34,45]
})
df
```

```python
df1=pd.DataFrame({
    'city':["faslabd","haiderbd","gujr"],
    'temp':[2,4,5]
    'hum':[23,35,56]
})
df1
```

```python
newdf=pd.concat([df,df1],ignore_index=True)
newdf
```

```python
#keys
newdf=pd.concat([df,df1],keys=['baray shehar','chhotay shehar'])
newdf
```

```python
#want to only single data frame
newdf.loc['baray shehar']
```
